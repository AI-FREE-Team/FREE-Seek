# STEP 1. 下載及載入模型 (請記得使用GPU)

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

# 確保使用的是 FREE-Seek 的模型
MODEL_NAME = "AI-FREE-Team/Llama-3.2-3B-Traditional-Chinese-FREE-Seek"

# 設定裝置 (使用 CUDA 或 CPU)
device = "cuda" if torch.cuda.is_available() else "cpu"

# 載入 Tokenizer 和模型
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
model = AutoModelForCausalLM.from_pretrained(
    MODEL_NAME,
    torch_dtype=torch.float16 if device == "cuda" else torch.float32,
    device_map="auto"
)

# LLaMA 3.2 Chat Template
def format_chat_prompt(system_prompt, user_prompt, history=[]):
    """
    格式化 chat template 以符合 LLaMA 3.2 3B 的格式

    :param system_prompt: 系統指示
    :param user_prompt: 使用者輸入
    :param history: 歷史對話 (list of tuples)
    :return: 格式化後的 prompt
    """
    prompt = "<|begin_of_text|><|start_header_id|>system<|end_header_id|>\n\n"
    prompt += f"{system_prompt}<|eot_id|><|start_header_id|>user<|end_header_id|>\n\n"

    for user_input, assistant_response in history:
        prompt += f"{user_input}<|eot_id|><|start_header_id|>assistant<|end_header_id|>\n{assistant_response}<|eot_id|><|start_header_id|>user<|end_header_id|>\n\n"

    prompt += f"{user_prompt}<|eot_id|><|start_header_id|>assistant<|end_header_id|>\n"
    return prompt
```

# 運行模型

```python
import time

start = time.time()
# 設定系統指示與對話
system_prompt = """ """

user_prompt = "請問肚子餓吃香蕉好嗎?"
history = []  # 歷史對話 (可選)

print("使用者問題: ", user_prompt)
print()
print("FREE-Seek: ", end="")

# Prompt 套用 llama3 格式
formatted_prompt = format_chat_prompt(system_prompt, user_prompt, history)

# Tokenization
inputs = tokenizer(formatted_prompt, return_tensors="pt").to(device)

# inference
output = model.generate(**inputs, max_length=4096, do_sample=True, temperature=0.7, eos_token_id=128009)

# Decode 輸出
response_text = tokenizer.decode(output[0], skip_special_tokens=True)
print(response_text)
print('推論時間: ', time.time() - start)
```
